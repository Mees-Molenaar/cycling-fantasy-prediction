```python
import pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup 
import requests
```

```python
predictions_file = Path(r"data/predictions.txt")
```

```python
df = pd.read_json(predictions_file)
df.head(5)
```

```python
wielerpoule_rennerlijst_url = r"https://wielerpoule.nl/rennerlijst.html"
```

```python
req = requests.get(wielerpoule_rennerlijst_url)
soup = BeautifulSoup(req.text, "lxml")
```

```python
renners_tabel = soup.find("table")
```

```python
special_chars = {
    "Č": "C",
    "é": "e",
    "Ö": "O",
    "ł": "l",
    "ë": "e",
    "î": "i",
    "Ä": "A",
    "ß": "S",
    "ø": "o",
    "É": "E",
    "Ø": "Ø",
    "Ü": "U",
    "Á": "A",
    "ó": "o",
    "Ņ": "N",
    "Š": "S",
    "Í": "I"}
```

```python
df["rider_name"].replace(special_chars, regex=True, inplace=True)
```

```python
df.head(5)
```

```python
runner_prices = {}

for ind, row in enumerate(renners_tabel.find_all("tr"), 0):
    if ind > 0:
        renner = row.find("td", attrs={"align": "left"}).text

        # Structuur moet even hetzelfde zijn als in df
        # ACHTERNAAM Voornaam
        processed_name = []

        for ind, name in enumerate(renner.split(), 0):
            insert_place = len(processed_name) - ind
            processed_name.insert(insert_place, name)

        processed_name[0] = processed_name[0].upper()
        name = " ".join(processed_name)

        prijs = row.find("td", attrs={"align": "right"}).text
        processed_prijs = int(prijs.split('.')[0])
        
        runner_prices[name] = processed_prijs
```
